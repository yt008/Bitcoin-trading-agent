```python
import pandas as pd
import numpy as np

def load_csv(filename):
    df = pd.read_csv(filename)
    df.dropna(how='any', inplace=True)
    df.reset_index(drop=True)
    # Convert the date column type from string to datetime for proper sorting.
    df['date'] = pd.to_datetime(df['date'])

    # Make sure historical prices are sorted chronologically, oldest first.
    df.sort_values(by='date', ascending=True, inplace=True)

    df.reset_index(drop=True, inplace=True)

    # Format timestamps as you want them to appear on the chart buy/sell marks.
    df['date'] = df['date'].dt.strftime('%Y-%m-%d %I:%M %p')

    # PlotlyTradingChart expects 'datetime' as the timestamps column name.
    df.rename(columns={'date': 'datetime'}, inplace=True)

    return df
```

```python
df = load_csv("../A2/BTCUSD_12h_2011-09-13_to_2019-10-23_bitstamp.csv")
df.head()
```

```python
import pickle
import copy
from keras.models import load_model
scaler = pickle.load(open('../A2/ZhongXianLuo_A1_scaler.pkl', 'rb'))
trend_scaler = pickle.load(open('../A2/ZhongXianLuo_A2_scaler.pkl', 'rb'))

def preprocess(files):
    for file in files:
        df = copy.deepcopy(file)
        df.dropna(how='any', inplace=True)
        df.reset_index(drop=True)
        df['date'] = pd.to_datetime(df['date'])
        df.drop(columns=['open', 'high', 'low'], inplace=True)

        for i in (1, 2, 3, 4, 5):
            temp_d = 'p_prev_' + str(i)
            temp_v = 'v_prev_' + str(i)
            df[temp_d] = df.close.shift(i)
            df[temp_v] = df.volume.shift(i)

        trend = pd.read_csv("~/trend.csv")
        trend['date'] = pd.to_datetime(trend['date'])
        trend = trend.loc[:,['date', 'bitcoin']]
        
        df.set_index('date', inplace=True)
        df = df.join(trend.set_index('date'), on='date')
        
        # df = df.join(trend.set_index('date'), on='date')
        #df = df.join(trend, on='date')
        for i in range(1, 5):
            tmp = 'bitcoin' + str(i)
            df[tmp] = df.bitcoin.shift(i)
            df[tmp].fillna(method='ffill', inplace=True)

        df.drop(columns=['bitcoin'], inplace=True)
        df.dropna(axis=0, inplace=True)
        trend = df[['bitcoin1', 'bitcoin2', 'bitcoin3', 'bitcoin4']]
        # df.drop(columns=['date', 'volume', 'bitcoin1', 'bitcoin2', 'bitcoin3', 'bitcoin4'], inplace=True)
        df.drop(columns=['volume', 'bitcoin1', 'bitcoin2', 'bitcoin3', 'bitcoin4'], inplace=True)
        
        names = df.columns
        indices = df.index.values.tolist()
        df = scaler.transform(df)
        trend = trend_scaler.transform(trend)

        X = df[:, 1:]
        y = df[:, 0]

        X = X.reshape(X.shape[0], 5, 2)

        return X, y, trend, names, indices


def model(model, X):
    model = load_model(model)
    result = model.predict(X)

    return result[:, 0]


def combiner(model, y1_hat, trend):
    model = load_model(model)
    train = np.concatenate((y1_hat.reshape(len(y1_hat), 1), trend), axis=1)
    train = train.reshape(len(train), train.shape[1], 1)
    result = model.predict(train)

    return result
```

```python
model_loc = "../A2/ZhongXianLuo_A1_model.h5"
cmodel = "../A2/ZhongXianLuo_A2_combiner.h5"
X, y, trend, names, indices = preprocess([price_history])
y1_hat = model(model_loc, X)
result = combiner(cmodel, y1_hat, trend)

X = X.reshape(X.shape[0], 10)
result = np.append(result, X, axis=1)
result = scaler.inverse_transform(result)
result = pd.DataFrame(result, index=indices, columns=names)
result.index = pd.to_datetime(result.index)
result = result[['close']]
```

```python
combine_features = dataset.join(result)
combine_features.close.shift(1)
combine_features.rename(columns={'close': 'next_pred'}, inplace=True)
combine_features.dropna(how='any', inplace=True)
```

```python
from tensortrade.exchanges import Exchange
from tensortrade.exchanges.services.execution.simulated import execute_order
from tensortrade.data import Stream, DataFeed, Module
from tensortrade.instruments import USD, BTC
from tensortrade.wallets import Wallet, Portfolio
from tensortrade.environments import TradingEnvironment

coinbase = Exchange("coinbase", service=execute_order)(
    Stream("USD-BTC", price_history['close'].tolist())
)

portfolio = Portfolio(USD, [
    Wallet(coinbase, 10000 * USD),
    Wallet(coinbase, 10 * BTC),
])

with Module("coinbase") as coinbase_ns:
    nodes = [Stream(name, dataset[name].tolist()) for name in dataset.columns]

feed = DataFeed([coinbase_ns])
feed.next()
```

```python
env = TradingEnvironment(
    feed=feed,
    portfolio=portfolio,
    action_scheme='managed-risk',
    reward_scheme='risk-adjusted',
    window_size=20
)
```

```python
from tensortrade.agents import A2CAgent
from tensortrade.agents import DQNAgent

agent = A2CAgent(env)
#agent = DQNAgent(env)
agent.train(n_episodes=2, n_steps=round(len(dataset)*0.8), save_path="../agents/")
agent.save(path='../agents/')
```
